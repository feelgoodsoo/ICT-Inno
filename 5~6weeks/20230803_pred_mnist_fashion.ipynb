```python
## rnn 으로 학습하여 정확도 0.9 이상 학습하고
## test[0] 번째 이미지를 예측하시요.

#20230803 task

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense, Input, Flatten, Conv2D, MaxPool2D,SimpleRNN, LSTM ,GRU,TimeDistributed
from tensorflow.keras import Sequential, Model
from tensorflow.keras.datasets.fashion_mnist import load_data
from tensorflow.keras.utils import to_categorical

(x_train,y_train),(x_test, y_test) = load_data()
# plt.imshow( x_train[0] ,cmap='gray')

## rnn으로 학습하여 정확도 0.9 이상 모델3, 4를 이용 테스트 0번째 이미지 예측

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# 정규화
x_train = x_train / 255
x_test = x_test / 255


class RNNModel(Model):
  #모델 객체 생성
  def __init__(self, shape_width, shape_length):
    super().__init__( Input(shape=(shape_width, shape_length)))
    self.lstm_layer = LSTM(128)
    self.hid_dense = Dense(64, name='hidden', activation='relu')
    self.out_dense = Dense(10, name='pred_img', activation='softmax')

  #모델 연결 및 훈련
  def call(self, input_x):
    print("--call method is invoked --")
    model = self.lstm_layer(input_x)
    model = self.hid_dense(model)
    model = self.out_dense(model)
    return model

# one-hot 인코딩
y_train = to_categorical(y_train)

# 모델 학습
#model = ModelLearning(x_train, y_train, 50, 32)
model = RNNModel(28,28)
model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['acc'])
model.fit(x_train, y_train, epochs=30, batch_size=30)

# 모델 예측
y_predict = model.predict(x_test)

# 모델 평가
label = class_names[y_predict[0].argmax()]
probability = y_predict[0][y_predict[0].argmax()]
print(label, probability)



```
